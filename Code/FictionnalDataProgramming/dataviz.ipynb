```python
# Importer les librairies ncessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("output_3.csv", sep="|")

# Paramtrage de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_colwidth', None)

# Afficher les premires lignes du dataset
print(df.head())

# Pie chart Nutri score : Rpartition des produits par nutriscore
plt.figure(figsize=(8,8))
plt.title('Rpartition des Nutriscores', size=20)
df['nutrition_grade_fr'].value_counts(normalize=True).sort_index().plot(kind='pie', autopct='%1.0f%%')
plt.axis('equal')
plt.show()

# Bar chart Categories : Rpartition des produits par catgories
plt.figure(figsize=(16,10))
plt.title('Rpartition des catgories (%)', size=20)
plt.xlabel('Catgories', size=15, fontweight='bold')
plt.ylabel('Proportion (%)', size=15, fontweight='bold')
df['renamed'].value_counts(normalize=True).plot(kind='bar')
plt.show()

# Slectionner uniquement les colonnes numriques
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Calculer la matrice de corrlation sur les colonnes numriques
correlation_matrix = df[numeric_cols].corr()

# Crer le heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.show()
```
